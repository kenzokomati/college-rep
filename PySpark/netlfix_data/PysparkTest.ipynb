# Quais colunas estão presentes no dataset?

```python
colunas = list(netflix_data.columns)

print(f"Colunas presentes no dataframe: {', '.join(colunas)}")
```

# Quantos filmes estão disponíveis na Netflix?

```python
# Agrupar por series e contar filmes/series
type_counts = netflix_data.groupby('type').size()

# Pegar numero de filmes
movie_count = type_counts['Movie']
print(f"Numero de filmes no netflix: {movie_count}")
```

# Quem são os 5 diretores com mais filmes e séries na plataforma?

```python
# Agrupar e contar diretores
director_counts = netflix_data.groupby('director').director.count()

# Ordenar os diretores em ordem decrescente
top_directors = director_counts.sort_values(ascending=False)

# Exibir os 5 diretores com mais filmes/séries
print(top_directors.head(5))
```

# Quais diretores também atuaram como atores em suas próprias produções?

```python
# Normalização
netflix_data['director'] = netflix_data['director'].fillna("Unknown")
netflix_data['cast'] = netflix_data['cast'].fillna("N/A")

directors = netflix_data['director']

directorsActors = []

for i in range(len(directors)):
  cast = netflix_data.iloc[i]['cast']

  # Checando se diretor está em elenco
  if directors[i] in cast:
      directorsActors.append(directors[i])

print(f"Diretores que também atuaram como atores em suas próprias produções: {', '.join(directorsActors)}")
```
